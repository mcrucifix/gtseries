```r
t <- seq(1024)
t_skewed <- t + rnorm(1024,sd=6.7)
plot(t,t_skewed, type='l')
harmonic_sample_data_skewed <- approx(t_skewed, harmonic_sample$data, t)
```

```r
harmonic_sample_data_skewed <- ts(harmonic_sample_data_skewed$y, deltat=1)
```
